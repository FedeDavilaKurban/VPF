## Quiero escribir archivos .dat en distintos Z para mandarle al Polaco

```python
from cicTools import *
from astropy.io import ascii
from astropy.table import Table

minmass = 1.

for snap in [67]:

    if minmass==1.: mass='10e11'

    filename = f'../data/TNG300-1_coord_{mass}_snap{snap}.dat'

    gxs = readTNG(minmass=1.,snap=snap)
    ascii.write(gxs, filename)
    print(f'Created {filename}')
```

```python
import matplotlib.pyplot as plt
from astropy.io import ascii

filenames = ['../data/TNG300-1_coord_10ala11.dat',\
             '../data/TNG300-1_coord_10e11_snap67.dat',\
             '../data/TNG300-1_coord_10e11_snap50.dat',\
             '../data/TNG300-1_coord_10e11_snap33.dat']


fig, axs = plt.subplots(1, 4, figsize=(15,3))

for filename,ax in zip(filenames,axs):
    gxs = ascii.read(filename)

    print(f'Mean interparticle separation: {205./(len(gxs))**(1./3)}')
    ax.hist2d(gxs['x']/1000.,gxs['y']/1000.,bins=128)
#    ax2.hist2d(gxs11['x'],gxs11['y'],bins=128)
#    ax3.hist2d(gxs12['x'],gxs12['y'],bins=128)
plt.show()
```
