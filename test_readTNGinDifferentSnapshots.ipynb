```python
import numpy as np
from cicTools import *
import matplotlib.pyplot as plt

gxs99 = readTNG(snap=99)
gxs50 = readTNG(snap=50)
gxs35 = readTNG(snap=35)
```

```python

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(6,10))
fig.suptitle('Snapshots 99, 50, and 35')

ax1.hist2d(gxs99['x'],gxs99['y'],bins=256)
ax2.hist2d(gxs50['x'],gxs50['y'],bins=256)
ax3.hist2d(gxs35['x'],gxs35['y'],bins=256)


plt.show()
```
